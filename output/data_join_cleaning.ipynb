```python
import pandas as pd
import os
import json
```

```python
folder_path = 'cleaning_data'  # Замените на путь к вашей папке

json_files = [f for f in os.listdir(folder_path) if f.endswith('.json')]
```

```python
# Пустой список для хранения данных из всех файлов
combined_data = []

# Чтение данных из каждого файла и добавление в общий список
for file in json_files:
    path = f'cleaning_data/{file}'
    with open(path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        combined_data.extend(data)

# Сохранение объединенных данных в новый файл
output_file = 'combined_data.json'
with open(output_file, 'w', encoding='utf-8') as f:
    json.dump(combined_data, f, ensure_ascii=False, indent=4)
```

```python
df_json = pd.read_json('combined_data.json')
df_json.head()
```

```python
folder_path = 'cleaning_data'  # Замените на путь к вашей папке

csv_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]

data_frames = []

# Загрузка данных из каждого файла и добавление в список

for file in csv_files:
    path = f'cleaning_data/{file}'
    data = pd.read_csv(path)
    data_frames.append(data)

# Объединение всех DataFrames в один
combined_data = pd.concat(data_frames, ignore_index=True)

# Сохранение объединенных данных в новый CSV файл
output_file = 'combined_data.csv'
combined_data.to_csv(output_file, index=False)
```
