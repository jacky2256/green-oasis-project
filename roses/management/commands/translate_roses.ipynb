```python
import pandas as pd
from googletrans import Translator
```

```python
df_roses = pd.read_csv('roses.csv')
```

```python
df_roses.head()
```

```python
df_roses['slug'] = df_roses['slug'].replace('%20', '_')
```

```python
# Преобразовать все значения в столбце 'slug' в маленькие буквы
df_roses['slug'] = df_roses['slug'].str.lower()
```

```python
# def translate(old_column, new_column, old_languages, new_lenguages):
#     translator = Translator()
#     translator.raise_Exception = True
#     df_roses[new_column] = df_roses[old_column].apply(lambda x: translator.translate(x, src=old_languages, dest=new_lenguages).text)
def translate(old_column, new_column, old_languages, new_languages):
    translator = Translator()
    translated_values = []

    for value in df_roses[old_column]:
        try:
            translated_text = translator.translate(value, src=old_languages, dest=new_languages).text
            translated_values.append(translated_text)
        except AttributeError as e:
            print(f"Error translating value '{value}' in column '{old_column}' to language '{new_languages}': {e}")
    
    df_roses[new_column] = translated_values
```
